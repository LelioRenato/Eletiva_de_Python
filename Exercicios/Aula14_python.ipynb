Adicionar um método `__str__` para que os objetos MeuTempo possam se imprimir decentemente.

````python
class MeuTempo:
    """Representa a hora do dia.
       atributos: hora, minuto, segundo
    """
    def __init__(self, hora=0, minuto=0, segundo=0):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo

    def __str__(self):
        """
        Retorna uma representação em string formatada do objeto MeuTempo.
        Formato: HH:MM:SS
        """
        # Formata os números com zeros à esquerda
        return f"{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}"

# Exemplo de Uso

# Instanciando um objeto MeuTempo
hora_atual = MeuTempo(10, 5, 30)

# Imprimindo o objeto diretamente
print(f"A hora atual é: {hora_atual}")
````

1. Escreva uma função booleana  chamada `entre`  que tome dois objetos MeuTempo, t1 e t2 , como argumentos, e retorne `True` se um terceiro objeto MeuTempo invocado estiver entre os dois tempos. Suponha que o tempo t1 <= t2 , e faça o teste fechado no limite inferior e abra no limite superior, isto é, retorne True se t1 <= obj < t2.

````python
class MeuTempo:
    """Representa a hora do dia.
       atributos: hora, minuto, segundo
    """
    def __init__(self, hora=0, minuto=0, segundo=0):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo

    def __str__(self):
        """Retorna uma representação em string formatada HH:MM:SS."""
        return f"{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}"

    def tempo_para_segundos(self):
        """
        Converte o objeto MeuTempo em um único valor inteiro de segundos totais
        desde a meia-noite, para facilitar a comparação.
        """
        minutos_totais = self.hora * 60 + self.minuto
        segundos_totais = minutos_totais * 60 + self.segundo
        return segundos_totais

    def entre(self, t1, t2):
        """
        Verifica se o objeto MeuTempo atual está entre t1 e t2,
        incluindo t1, mas não incluindo t2 (t1 <= self < t2).
        Assume que t1 <= t2.
        """
        # Converte todos os tempos para segundos totais para uma comparação simples
        segundos_self = self.tempo_para_segundos()
        segundos_t1 = t1.tempo_para_segundos()
        segundos_t2 = t2.tempo_para_segundos()

        # Retorna True se a condição fechado/aberto for atendida
        return segundos_t1 <= segundos_self < segundos_t2

# Exemplo de Uso

t1 = MeuTempo(10, 0, 0)  # (Início do intervalo)
t2 = MeuTempo(12, 0, 0)  # (Fim do intervalo)

print(f"Intervalo de referência: [{t1} a {t2})")

obj1 = MeuTempo(10, 0, 0)
print(f"Objeto 1 ({obj1}) está entre t1 e t2? {obj1.entre(t1, t2)}")

obj2 = MeuTempo(11, 30, 0)
print(f"Objeto 2 ({obj2}) está entre t1 e t2? {obj2.entre(t1, t2)}")

obj3 = MeuTempo(12, 0, 0)
print(f"Objeto 3 ({obj3}) está entre t1 e t2? {obj3.entre(t1, t2)}")

obj4 = MeuTempo(9, 59, 59)
print(f"Objeto 4 ({obj4}) está entre t1 e t2? {obj4.entre(t1, t2)}")
````

3. Sobrecarregue o(s) operador(es) necessário(s)  --ver a lista de métodos especiais em https://docs.python.org/3/reference/datamodel.html#special-method-names -- para que, em vez de ter que escrever :
```python
if t1.depois( t2 ):  
     ...
```
possamos usar o mais conveniente:
```python
if t1  >  t2 :  
     ...
```

````python
class MeuTempo:
    """Representa a hora do dia e suporta comparação de operadores."""
    def __init__(self, hora=0, minuto=0, segundo=0):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo

    def __str__(self):
        return f"{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}"

    def tempo_para_segundos(self):
        """Converte o objeto MeuTempo em segundos totais desde a meia-noite."""
        return self.hora * 3600 + self.minuto * 60 + self.segundo

    # ---------------------------------------------------
    # Sobrecarga de Operadores de Comparação
    # ---------------------------------------------------

    def __eq__(self, other):
        """Sobrecarga do operador de igualdade: t1 == t2"""
        return self.tempo_para_segundos() == other.tempo_para_segundos()

    def __ne__(self, other):
        """Sobrecarga do operador de diferença: t1 != t2"""
        return self.tempo_para_segundos() != other.tempo_para_segundos()

    def __lt__(self, other):
        """Sobrecarga do operador 'menor que': t1 < t2"""
        return self.tempo_para_segundos() < other.tempo_para_segundos()

    def __le__(self, other):
        """Sobrecarga do operador 'menor ou igual que': t1 <= t2"""
        return self.tempo_para_segundos() <= other.tempo_para_segundos()

    def __gt__(self, other):
        """Sobrecarga do operador 'maior que': t1 > t2"""
        return self.tempo_para_segundos() > other.tempo_para_segundos()

    def __ge__(self, other):
        """Sobrecarga do operador 'maior ou igual que': t1 >= t2"""
        return self.tempo_para_segundos() >= other.tempo_para_segundos()

# Exemplo de Uso

t1 = MeuTempo(10, 30, 0)
t2 = MeuTempo(12, 0, 0)
t3 = MeuTempo(10, 30, 0)

print(f"t1: {t1}, t2: {t2}, t3: {t3}")

# Agora podemos usar os operadores de comparação diretamente:

print(f"\nUsando operadores sobrecarregados:")
print(f"t1 > t2 ? {t1 > t2}")
print(f"t1 < t2 ? {t1 < t2}")
print(f"t2 >= t1 ? {t2 >= t1}")
````

Defina um novo tipo de tartaruga, TurtleGTX, que venha com alguns recursos extras: ela pode saltar para uma determinada distância, e tem um odômetro que rastreia até onde a tartaruga viajou desde que saiu da linha de produção. (A classe base tem vários sinônimos como `fd` e `forward` ; `back` , `backward` e `bk` : para este exercício, concentre-se **apenas** em colocar essa funcionalidade no método `forward`.) Pense cuidadosamente sobre como contar a distância se a tartaruga for solicitada a avançar por um valor negativo. (Não gostaríamos de comprar uma tartaruga de segunda mão cuja leitura do odômetro foi falsificada porque seu proprietário anterior a levou para trás ao redor do quarteirão com muita frequência. Tente isso em um carro perto de você e veja se o odômetro do carro conta para cima ou para baixo quando você marcha ré.)

````python
import turtle
import math

class TurtleGTX(turtle.Turtle):
    """
    Um novo tipo de tartaruga com recursos extras:
    - Um odômetro que rastreia a distância total viajada.
    - Um método jump() para saltos sem desenhar.
    """
    def __init__(self):
        # Inicializa a classe base Turtle
        super().__init__()
        # Inicializa o novo atributo odometro (odometer)
        self.odometer = 0.0
        self.speed(1) # Define uma velocidade padrão

    def forward(self, distance):
        """
        Move a tartaruga para frente e atualiza o odômetro.
        Conta a distância absoluta percorrida.
        """
        # Chama o método forward() da classe base Turtle
        super().forward(distance)

        # O odômetro rastreia a distância absoluta percorrida
        # abs(distance) garante que mover para trás (distância negativa)
        # ainda aumente a leitura do odômetro.
        self.odometer += abs(distance)

    def jump(self, distance, angle=0):
        """
        Salto: Move a tartaruga sem desenhar e sem contar no odômetro.
        """
        self.penup()
        self.right(angle)
        super().forward(distance) # Usa o forward da classe base para não contar no odômetro
        self.left(angle)
        self.pendown()

    def get_odometer_reading(self):
        """
        Retorna a leitura atual do odômetro.
        """
        return self.odometer

#  Demonstração do Uso

if __name__ == '__main__':
    jn = turtle.Screen()
    jn.bgcolor("white")

    # Instancia nossa nova tartaruga TurtleGTX
    ninja = TurtleGTX()
    ninja.color("red")
    ninja.pensize(3)

    print(f"Leitura inicial do odômetro: {ninja.get_odometer_reading():.2f}")

    # Movimento para frente
    ninja.forward(100)
    print(f"Após 100 para frente: {ninja.get_odometer_reading():.2f}")

    # Movimento para trás (valor negativo)
    ninja.left(90)
    ninja.forward(-50)
    print(f"Após -50 (para trás): {ninja.get_odometer_reading():.2f}") # O odômetro aumentou para 150

    # Usando o salto (não deve aumentar o odômetro)
    ninja.jump(20, 90)
    ninja.forward(30)
    print(f"Após salto e 30 para frente: {ninja.get_odometer_reading():.2f}") # O odômetro aumentou apenas de 150 para 180

    ninja.hideturtle()
    jn.mainloop()
````
