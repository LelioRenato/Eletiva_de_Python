2. Escreva um conjunto apropriadamente geral de funções que podem desenhar formas como na figura:

```python
import turtle, math
ninja = turtle.Turtle()

def isosceles(t, leg, angulo_base, base):

    t.fd(leg)
    t.lt(180 + angulo_base)
    t.fd(base)
    t.lt(180 + angulo_base)
    t.fd(leg)

def polypie(t, n, leg):
    vertex_angle = 360/n
    angulo_base = (180 - vertex_angle)/2
    base = math.sqrt(2 * (leg**2) - ((2 * (leg**2)) * math.cos(vertex_angle * (math.pi/180))))
    for i in range(n):
        isosceles(t, leg, angulo_base, base)
        t.rt(180 - 2 * vertex_angle)

polypie(ninja, 7, 100)
turtle.mainloop()
```

3. Adicione instruções para desenhar o talo e folha da flor. Faça a refatoração apropriada para a interface da flor completa ficar "limpa".

```python
import turtle
import math
ninja = turtle.Turtle()

#Flor - Exercicio 1
def petala(t, raio, angulo_petala):
    """Desenha uma única pétala de flor."""
    # Desenha um semicírculo
    t.circle(raio, angulo_petala)
    # Volta ao centro e desenha a outra metade da pétala
    t.left(180 - angulo_petala)
    t.circle(raio, angulo_petala)
    # Retorna à posição inicial para desenhar a próxima pétala
    t.left(180 - angulo_petala)

def flor(t, num_petalas, raio_petala, angulo_petala):
    """Desenha uma flor com um número especificado de pétalas."""
    angulo_entre_petalas = 360 / num_petalas
    for _ in range(num_petalas):
        petala(t, raio_petala, angulo_petala)
        t.left(angulo_entre_petalas)


# Exercício 3: talo + folha
def talo_e_folha(t, comprimento=200, raio_folha=50):
    """Desenha um talo com uma folha."""
    t.pensize(5)
    t.color("green")

    # Talo
    t.right(90)     # girar para baixo
    t.forward(comprimento)

    # Folha (duas semicircunferências opostas)
    t.left(45)
    t.begin_fill()
    t.circle(raio_folha, 90)
    t.left(90)
    t.circle(raio_folha, 90)
    t.end_fill()

#Exercício 1

#Desenhando uma flor 1?
ninja.reset()
ninja.speed(10)
petala(ninja, 100, 60)

ninja.speed(10)
flor(ninja, 6, 100, 60)

#Flor 2
ninja.reset()
ninja.speed(10)
petala(ninja, 100, 60)

ninja.speed(10)
flor(ninja, 10, 100, 60)

#Flor 3
ninja.reset()
ninja.speed(10)
petala(ninja, 100, 60)

#ninja.speed(10)
flor(ninja, 20, 100, 60)


# Flor com talo
ninja.reset()
ninja.speed(10)

# Desenha a flor
flor(ninja, 10, 100, 60)

# Vai para o centro da flor
ninja.penup()
ninja.home()
ninja.pendown()

# Desenha o talo e a folha
talo_e_folha(ninja, comprimento=100, raio_folha=30)

ninja.hideturtle()
turtle.done()
```

Você pode obter uma solução em http://thinkpython2.com/code/letters.py; também requer http://thinkpython2.com/code/polygon.py.

```python
import turtle
import math

# Funções Auxiliares Básicas

def move_to(t, x, y):
    """Move a tartaruga para a coordenada (x, y) sem desenhar."""
    t.penup()
    t.goto(x, y)
    t.pendown()

def draw_line(t, length):
    """Desenha uma linha para frente e retorna à posição inicial."""
    t.forward(length)
    t.backward(length)

def draw_arc(t, r, angle):
    """Desenha um arco de um círculo de raio 'r' com o ângulo especificado."""
    passos_por_circulo = 50
    n_passos = int(passos_por_circulo * angle / 360)
    circunferencia_total = 2 * math.pi * r
    length = circunferencia_total / passos_por_circulo
    angulo_por_passo = float(angle) / n_passos
    for _ in range(n_passos):
        t.forward(length)
        t.left(angulo_por_passo)

# Funções de Desenho das Letras
# As letras são desenhadas assumindo um tamanho de altura padrão (ex: 100 unidades)

def draw_a(t, height=100):
    """Desenha a letra A."""
    width = height * 0.6
    t.left(75)
    t.forward(height)
    t.right(150)
    t.forward(height)
    t.backward(height / 2)
    t.right(75) # Vira para a direita para desenhar a barra horizontal
    t.forward(width * 0.5)
    t.backward(width * 0.5)
    t.left(75) # Retorna à orientação original da barra horizontal
    t.forward(height / 2) # Retorna ao ponto central inferior
    t.left(75) # Retorna à orientação inicial

def draw_b(t, height=100):
    """Desenha a letra B."""
    t.forward(height)
    t.backward(height / 2)
    t.right(90)
    draw_arc(t, height/4, 180) # Semicírculo superior
    t.left(180) # Vira para desenhar o arco inferior
    draw_arc(t, height/4, 180) # Semicírculo inferior
    t.left(90) # Retorna à orientação inicial

def draw_c(t, height=100):
    """Desenha a letra C."""
    t.penup()
    t.forward(height/2)
    t.pendown()
    t.left(90)
    draw_arc(t, height/2, 180)
    t.left(90)
    t.forward(height/2) # Fecha a base inferior do C
    t.penup()
    t.backward(height/2)
    t.left(90)
    t.forward(height)
    t.right(90)
    t.backward(height/2)
    t.pendown()


def draw_d(t, height=100):
    """Desenha a letra D."""
    t.forward(height)
    t.backward(height)
    t.right(90)
    draw_arc(t, height/2, 180)
    t.left(90)

def draw_e(t, height=100):
    """Desenha a letra E."""
    width = height * 0.6
    t.forward(height)
    t.right(90)
    t.forward(width)
    t.backward(width)
    t.left(90)
    t.backward(height / 2)
    t.right(90)
    t.forward(width * 0.75) # Barra do meio
    t.backward(width * 0.75)
    t.left(90)
    t.backward(height / 2)
    t.right(90)
    t.forward(width) # Barra inferior
    t.backward(width)
    t.left(90)

# Testando

if __name__ == '__main__':
    jn = turtle.Screen()
    joana = turtle.Turtle()
    joana.speed(1)
    joana.pensize(5)
    joana.color("darkgreen")

    # Desenha as letras A, B, C, D, E em sequência
    draw_a(joana, 100)
    move_to(joana, 40, 0) # Move para a próxima posição inicial (ajuste manual)

    draw_b(joana, 100)
    move_to(joana, 110, 0)

    draw_c(joana, 100)
    move_to(joana, 200, 0)

    draw_d(joana, 100)
    move_to(joana, 280, 0)

    draw_e(joana, 100)

    joana.hideturtle() # Esconde a tartaruga no final
    jn.mainloop()
```

5. Leia sobre espirais em http://en.wikipedia.org/wiki/Spiral; em seguida, escreva um programa que desenhe uma espiral arquimediana (ou um dos outros tipos).

```python
import turtle
ninja = turtle.Turtle()

def spiral(t, n, length = 3, a = 0.1,
            b = 0.0002):

    theta = 0.0
    for i in range(n):
        t.fd(length)
        dtheta = 1 / (a + b * theta)
        t.lt(dtheta)
        theta += dtheta

spiral(ninja, 1000, 2, 0.2, 0.002)
turtle.mainloop()
```
