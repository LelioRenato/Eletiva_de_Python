Exercício 6:

Sabendo o ângulo interno da ponta de uma estrela, desenhe quatro estrelas em uma janela, com uma certa distância entre elas. Dica: use a função penup() e pendown() do módulo turtle

```python
import turtle

# Configura a janela (jn)
jn = turtle.Screen()
jn.bgcolor("lightblue")
jn.title("Quatro Estrelas")

# Cria a tartaruga
ninja = turtle.Turtle()
ninja.shape("turtle")
ninja.color("gold")
ninja.pensize(2)
ninja.speed(0)  # Definindo para a velocidade máxima para desenhar rapidamente

def desenhar_estrela(tamanho):
    """Desenha uma estrela de 5 pontas."""
    ninja.pendown()
    for _ in range(5):
        ninja.forward(tamanho)
        ninja.right(144)
    ninja.penup()

# Desenhando as quatros estrelas

# Primeira estrela
ninja.penup()
ninja.goto(-200, 50) # Move para a posição inicial (superior esquerda)
desenhar_estrela(100) # Desenhando uma estrela de 100 unidades

# Segunda estrela
ninja.goto(-50, 50) # Move para a próxima posição
desenhar_estrela(80)  # Desenhando uma estrela menor

# Terceira estrela
ninja.goto(100, 50)  # Move para a próxima posição
desenhar_estrela(60)  # Desenhando outra estrela

# Quarta estrela
ninja.goto(220, 50) # Move para a última posição
desenhar_estrela(40)  # Desenhando a menor estrela

# Move a tartaruga para uma posição final para que ela não fique em cima da última estrela
ninja.goto(0, -100)
ninja.color("black")
ninja.write("Quatro Estrelas Desenhadas! Hiaaaa", align="center", font=("Arial", 16, "normal"))

jn.mainloop()
```

3. Faça uma cópia da função square e mude o nome para polygon. Adicione um outro parâmetro chamado n e modifique o corpo da função para que ela desenhe um polígono regular de n lados. Dica: Os ângulos externos de um polígono regular de n lados são de 360/n graus.

```python
import turtle

def polygon(t, length, n):
    """
    Desenha um polígono regular de 'n' lados usando o objeto Turtle (t),
    com lados do comprimento especificado (length).
    """
    angulo_externo = 360 / n
    for _ in range(n):
        t.forward(length)
        t.left(angulo_externo)

jn = turtle.Screen()
jn.bgcolor("white")

joana = turtle.Turtle()
joana.pensize(2)
joana.speed(3)

# Desenhando um triângulo (n=3)
joana.color("red")
polygon(joana, 80, 3)

# Mover para a próxima posição para não sobrepor o desenho
joana.penup()
joana.goto(0, 100)
joana.pendown()

# Desenhando um quadrado (n=4)
joana.color("blue")
polygon(joana, 80, 4)

# Movendo novamente
joana.penup()
joana.goto(100, 0)
joana.pendown()

# Desenhando um hexágono (n=6)
joana.color("green")
polygon(joana, 50, 6)

# Movendo novamente
joana.penup()
joana.goto(-100, -100)
joana.pendown()

# Desenhando um octógono (n=8)
joana.color("purple")
polygon(joana, 40, 8)

jn.mainloop()
```

4. Escreva uma função chamada circle que toma como argumento uma Turtle, t, e um raio, r, e desenha um círculo aproximado ao chamar a função polygon com length e número n de lados apropriados. Teste sua função com um intervalo de valores de r. Dica: Determine a circunferência do círculo e se certifique que comprimento * n = circunferência.



```python
import turtle
import math

def polygon(t, length, n):
    """
    Desenha um polígono regular de 'n' lados.
    """
    angulo_externo = 360 / n
    for _ in range(n):
        t.forward(length)
        t.left(angulo_externo)

def circle(t, r):
    """
    Desenha um círculo aproximado de raio 'r' usando a função polygon.
    """
    # Usamos um número grande de lados (n) para uma aproximação mais suave
    n = 50
    # Calculamos o comprimento de cada lado para que o perímetro total seja a circunferência (2 * pi * r)
    circunferencia = 2 * math.pi * r
    length = circunferencia / n

    # Chamando a função polygon para desenhar o círculo
    polygon(t, length, n)

jn = turtle.Screen()
jn.bgcolor("white")
joana = turtle.Turtle()
joana.speed(5) # Aumenta a velocidade para ver os testes mais rapidamente
joana.pensize(2)

# Teste 1: Raio 30
joana.color("red")
circle(joana, 30)

# Mover para a próxima posição
joana.penup()
joana.goto(100, 0)
joana.pendown()

# Teste 2: Raio 60
joana.color("blue")
circle(joana, 60)

# Mover para a próxima posição
joana.penup()
joana.goto(-150, 50)
joana.pendown()

# Teste 3: Raio 80 (círculo maior)
joana.color("green")
circle(joana, 80)

jn.mainloop()
```

5. Faça uma versão mais geral da função circle, chamada arc, que toma um parâmetro adicional, angle, que determina qual a fração do circulo a ser desenhada. angle está em unidades de graus, logo quando angle=360, a função arc deve desenhar um cículo completo.

```python
import turtle
import math

def arc(t, r, angle):
    """
    Desenha um arco de um círculo de raio 'r' com o ângulo especificado (em graus).
    """
    # Usamos 50 passos por círculo completo (360 graus) para mais suavidade
    passos_por_circulo = 50

    # Calcula o número total de passos necessários para o ângulo fornecido
    n_passos = int(passos_por_circulo * angle / 360)

    # Calcula o comprimento de cada passo (com base na circunferência total)
    circunferencia_total = 2 * math.pi * r
    length = circunferencia_total / passos_por_circulo

    # Calcula o ângulo de viragem em cada passo
    angulo_por_passo = float(angle) / n_passos

    # Desenha o arco
    for _ in range(n_passos):
        t.forward(length)
        t.left(angulo_por_passo)

jn = turtle.Screen()
jn.bgcolor("white")
joana = turtle.Turtle()
joana.pensize(3)
joana.speed(5)

# Teste 1: Um semicírculo (ângulo = 180)
joana.color("red")
arc(joana, 80, 180)

# Mover para a próxima posição
joana.penup()
joana.goto(100, 0)
joana.pendown()

# Teste 2: Um quarto de círculo (ângulo = 90)
joana.color("blue")
arc(joana, 80, 90)

# Mover para a próxima posição
joana.penup()
joana.goto(0, -100)
joana.pendown()
joana.right(90) # Ajusta a direção para começar o próximo arco virado para cima

# Teste 3: Um círculo completo (ângulo = 360) - igual à função circle()
joana.color("green")
arc(joana, 50, 360)

jn.mainloop()
```
