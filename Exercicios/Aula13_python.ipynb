6. Escreva uma definição para uma classe chamada **Circulo** com atributos ***centro*** e ***raio***, em que ***centro*** é um objeto do tipo **Ponto** e ***raio*** é um número.
   * Instancie um objeto **Circulo** que represente um círculo com seu centro em (150, 100) e raio 75 e imprima na tela as coordenadas (x,y) do centro do círculo.
   * Escreva uma função chamada ponto_no_circulo que use um Circulo e um Ponto e retorne True se o Ponto estiver dentro do círculo ou no limite do mesmo.

```python
import math

# Definição da classe Ponto
class Ponto:
    """Representa um ponto no espaço 2D."""
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        """Método para uma representação amigável ao imprimir o objeto."""
        return f"({self.x}, {self.y})"

# Definição da classe Circulo
class Circulo:
    """
    Representa um círculo.

    Atributos:
    - centro: do tipo Ponto.
    - raio: do tipo número (int ou float).
    """
    def __init__(self, centro, raio):
        self.centro = centro
        self.raio = raio

    def __str__(self):
        """Representação em string amigável para o objeto Circulo."""
        return f"Círculo: Centro={self.centro}, Raio={self.raio}"

# Instanciando e Imprimindo o Círculo

# Criar o objeto Ponto para o centro (150, 100)
centro_circulo = Ponto(150, 100)

# Criar o objeto Circulo com centro em (150, 100) e raio 75
meu_circulo = Circulo(centro=centro_circulo, raio=75)

# Imprimir na tela as coordenadas (x,y) do centro do círculo
print(f"O centro do círculo está nas coordenadas: ({meu_circulo.centro.x}, {meu_circulo.centro.y})")
print("-" * 40)

# Função ponto_no_circulo

def ponto_no_circulo(circulo_obj, ponto_obj):
    """
    Verifica se um Ponto está dentro ou no limite de um Circulo.

    Args:
        circulo_obj (Circulo): O objeto círculo de referência.
        ponto_obj (Ponto): O objeto ponto a ser verificado.

    Returns:
        bool: True se o ponto estiver dentro/no limite, False caso contrário.
    """
    # Calcula a distância entre o centro do círculo e o ponto usando a fórmula da distância euclidiana
    distancia = math.sqrt(
        (circulo_obj.centro.x - ponto_obj.x)**2 +
        (circulo_obj.centro.y - ponto_obj.y)**2
    )

    # Retorna True se a distância for menor ou igual ao raio
    return distancia <= circulo_obj.raio

# Exemplo de Uso

# Ponto de teste 1: Dentro do círculo
ponto_teste_dentro = Ponto(150, 100)
esta_dentro_1 = ponto_no_circulo(meu_circulo, ponto_teste_dentro)
print(f"O ponto {ponto_teste_dentro} está no círculo? {esta_dentro_1}")

# Ponto de teste 2: Fora do círculo
ponto_teste_fora = Ponto(0, 0)
esta_dentro_2 = ponto_no_circulo(meu_circulo, ponto_teste_fora)
print(f"O ponto {ponto_teste_fora} está no círculo? {esta_dentro_2}")

# Ponto de teste 3: No limite do círculo
ponto_teste_limite = Ponto(225, 100)
esta_dentro_3 = ponto_no_circulo(meu_circulo, ponto_teste_limite)
print(f"O ponto {ponto_teste_limite} está no círculo? {esta_dentro_3}")
```

7. Escreva uma função chamada **desenhar_ret** que use um objeto ***Turtle*** e um ***Retangulo*** e use o Turtle para desenhar o retângulo.

```python
import turtle

# Definição da classe Ponto
class Ponto:
    """Representa um ponto no espaço 2D."""
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"({self.x}, {self.y})"

# Definição da classe Retangulo
class Retangulo:
    """
    Representa um retangulo.

    Atributos:
    - largura: do tipo float.
    - altura: do tipo float.
    - canto: do tipo Ponto (representa o canto inferior esquerdo).
    """
    def __init__(self, largura, altura, canto):
        self.largura = largura
        self.altura = altura
        self.canto = canto

    def __str__(self):
        return (f"Retangulo: Largura={self.largura}, Altura={self.altura}, "
                f"Canto Inferior Esquerdo={self.canto}")

def desenhar_ret(tarta, ret):
    """
    Usa um objeto Turtle (tarta) para desenhar um Retangulo (ret).
    """
    # Mover a tartaruga para o canto inferior esquerdo do retângulo
    tarta.penup()
    tarta.goto(ret.canto.x, ret.canto.y)
    tarta.pendown()

    # Desenhar os 4 lados do retângulo

    # Lado inferior (largura)
    tarta.forward(ret.largura)
    tarta.left(90) # Vira para cima

    # Lado direito (altura)
    tarta.forward(ret.altura)
    tarta.left(90) # Vira para a esquerda

    # Lado superior (largura)
    tarta.forward(ret.largura)
    tarta.left(90) # Vira para baixo

    # Lado esquerdo (altura) - fechando a forma
    tarta.forward(ret.altura)
    tarta.left(90) # Retorna à orientação original (para a direita)

# Exemplo de Uso

# Configuração da janela Turtle
jn = turtle.Screen()
jn.bgcolor("white")

# Criação do objeto Turtle
dragon = turtle.Turtle()
dragon.pensize(3)
dragon.color("red")
dragon.speed(1) # Velocidade lenta para ver o desenho

# Criação do objeto Ponto para o canto do retângulo (inicia em x=0, y=0)
canto_inferior = Ponto(0, 0)

# Criação do objeto Retangulo
meu_ret = Retangulo(largura=150, altura=80, canto=canto_inferior)

# Chamada da função que desenha o retângulo
desenhar_ret(dragon, meu_ret)

# Mantém a janela aberta até o usuário fechar
jn.mainloop()
```

8. Escreva uma função chamada desenhar_circulo que use uma ***Turtle*** e um ***Circulo*** e desenhe o ***Circulo***.

```python
import turtle
import math

# Definição da classe Ponto
class Ponto:
    """Representa um ponto no espaço 2D."""
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"({self.x}, {self.y})"

# Definição da classe Circulo
class Circulo:
    """
    Representa um círculo.

    Atributos:
    - centro: do tipo Ponto.
    - raio: do tipo número (int ou float).
    """
    def __init__(self, centro, raio):
        self.centro = centro
        self.raio = raio

    def __str__(self):
        return f"Círculo: Centro={self.centro}, Raio={self.raio}"

def desenhar_circulo(tarta, circulo_obj):
    """
    Usa um objeto Turtle (tarta) para desenhar um objeto Circulo (circulo_obj).
    """
    # Calcular a posição inicial: o método circle() do Turtle desenha do ponto mais baixo do círculo (tangente inferior).
    # Precisamos mover a tartaruga para: (centro.x, centro.y - raio).
    pos_x = circulo_obj.centro.x
    pos_y = circulo_obj.centro.y - circulo_obj.raio

    tarta.penup()
    tarta.goto(pos_x, pos_y)
    tarta.pendown()

    # Desenhar o círculo usando o método embutido do Turtle
    tarta.circle(circulo_obj.raio)

# Exemplo de Uso

# Configuração da janela Turtle
jn = turtle.Screen()
jn.bgcolor("white")

# Criação do objeto Turtle
caveira = turtle.Turtle()
caveira.pensize(3)
caveira.color("blue")
caveira.speed(2) # Velocidade de desenho

# Criação do objeto Ponto para o centro do círculo
centro_exemplo = Ponto(50, 50)

# Criação do objeto Circulo com raio
meu_circulo = Circulo(centro=centro_exemplo, raio=75)

# Chamada da função que desenha o círculo
desenhar_circulo(caveira, meu_circulo)

# Mantém a janela aberta até o usuário fechar
jn.mainloop()
```
