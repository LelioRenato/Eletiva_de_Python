#Criando módulos

Tudo o que precisamos para criar um módulo é um arquivo de texto com uma extensão .py no nome do arquivo:

```python
# esse comando logo abaixo é do jupyter notebook para escrever um arquivo .py
#a dupla porcentagem é para colocar um comando que não está no python
%%writefile Seqtools.py
def  remover_em ( pos ,  seq ):
    return  seq [: pos ]  +  seq [ pos + 1 :]
```

```python
# ler o aqruivo que acabamos de criar:
%cat Seqtools.py
```

Agora podemos usar nosso módulo nos scripts e no shell do Python. Para fazer isso, devemos primeiro importar o módulo. Existem duas maneiras de fazer isso:

```python
import  Seqtools
s  =  "O string!"
Seqtools.remover_em( 7 ,  s ) #ele vai até o sexto e imprimir, porém o setimo ele não ira imprimir (vai parar)
```

No primeiro exemplo o nome do módulo e um ponto é escrito antes do nome da função, separados de um ponto(.). No segundo exemplo só a função remover_em é importada, e a chamada é feita exatamente como as funções que vimos anteriormente, sem necesidade de incluir o nome do módulo.

Observe que, em ambos os casos, não incluímos a extensão do arquivo .py ao importar. O Python espera que os nomes de arquivos dos módulos do Python sejam finalizados em .py , portanto, a extensão do arquivo não é incluída na instrução de importação.

O uso de módulos permite dividir programas muito grandes em partes de tamanho gerenciável e manter as partes relacionadas juntas.

Outro exemplo é apresentado com o arquivo lcount.py definido assim:


```python
%%writefile lcount.py

def linecount(filename):
    count = 0
    for line in open(filename):
        count += 1
    return count

linecount('lcount.py')
```

```python
%cat lcount.py
```

Programas que serão importados como módulos geralmente usam o seguinte idioma:

```python
if __name__ == '__main__':
    print(linecount('lcount.py'))
```

Pontos importantes:

    O identificador __name__ é uma variável interna que é configurada quando o programa é iniciado;

    Se o programa estiver sendo executado como um script, __name__ tem o valor '__main__';
        Nesse caso, o código de teste é executado.

    Caso contrário, se o módulo estiver sendo importado, o código de teste será ignorado.

# Capturando exceções

Para evitar esses erros, você poderia usar funções como os.path.exists e os.path.isfile, mas levaria muito tempo e código para verificar todas as possibilidades (se “Errno 21” for qualquer indicação, há pelo menos 21 coisas que podem dar errado).

É melhor fazer uma tentativa e lidar com possíveis problemas. A declaração try faz isso!!!!


```python
try:
    fin = open('bad_file')
except:
    print('Algo de errado aconteceu.')
```



    O Python começa executando a cláusula try.
        Se tudo correr bem:
            ele pula a cláusula except e continua.
        Se ocorrer uma exceção:
            ele sai da cláusula try e executa a cláusula except.

    Manipular uma exceção com uma instrução try é chamado de captura de uma exceção.

Para leitura sobre tipos de erro e a sua manipulação, sugerimos o tutorial https://docs.python.org/3/tutorial/errors.html

Uma forma de obter mais informação sobre o erro é chamar a excepção com um identificador (nesse caso err) e imprimi-lo assim como seu tipo.


```python
try:
    fin = open("another_bad_file")
except Exception as err:
    print(type(err))
    print(err)
```

```python
import os

def listar_arquivos(diretorio): #Lista todos os arquivos e subdiretórios em um dado diretório.
  for root, dirs, files in os.walk(diretorio):
    print(f"Diretório: {root}")
    for name in files:
      print(f"  Arquivo: {os.path.join(root, name)}")
    for name in dirs:
      print(f"  Subdiretório: {os.path.join(root, name)}")

# Diretório atual
cwd = os.getcwd()

# Listando os arquivos e subdiretórios
listar_arquivos(cwd)
```

5. Adapte o script acima para ser um módulo.

```python
import csv
import os

def ler_arquivos_com_delimiter(nome_arquivo, delimiter='\t'):
    """
    Lê dados de um arquivo com múltiplas colunas e retorna um dicionário,
    permitindo especificar o delimitador.

    Args:
        nome_arquivo: O nome do arquivo a ser lido.
        delimiter: O caractere usado para separar as colunas.

    Returns:
        dict: Um dicionário onde as chaves são os nomes das colunas (se houver cabeçalho)
              ou os números das colunas (começando em 0), e os valores são listas
              contendo os dados de cada coluna. Retorna um dicionário vazio em caso de erro.
    """
    dados = {}

    # Adicionando uma verificação de existência de arquivo mais robusta
    if not os.path.exists(nome_arquivo):
        print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado.")
        return dados

    try:
        with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:
            # Tenta detectar se há cabeçalho
            try:
                # O sniffer precisa de uma amostra do arquivo
                amostra = arquivo.read(1024)
                arquivo.seek(0) # Retorna ao início do arquivo
                tem_cabecalho = csv.Sniffer().has_header(amostra)
            except:
                # Pega o potencial de parada de interação ou outros erros
                tem_cabecalho = False
                arquivo.seek(0)

            reader = csv.reader(arquivo, delimiter=delimiter)

            # Lógica para processar o cabeçalho ou a primeira linha
            try:
                if tem_cabecalho:
                    cabecalho = next(reader)
                    for col_nome in cabecalho:
                        dados[col_nome] = []
                    num_colunas = len(cabecalho)
                else:
                    primeira_linha = next(reader)
                    # Reinicia o leitor para incluir a primeira linha nos dados
                    arquivo.seek(0)
                    reader = csv.reader(arquivo, delimiter=delimiter)
                    num_colunas = len(primeira_linha)
                    for i in range(num_colunas):
                        dados[i] = []
            except StopIteration:
                return {} # Arquivo vazio

            # Processa o restante das linhas
            for linha in reader:
                if len(linha) == num_colunas:
                    if tem_cabecalho:
                        for i, valor in enumerate(linha):
                            dados[cabecalho[i]].append(valor)
                    else:
                        for i, valor in enumerate(linha):
                            dados[i].append(valor)
                # Ignora linhas que não correspondem ao número de colunas esperado

    except Exception as e:
        print(f"Ocorreu um erro ao ler o arquivo '{nome_arquivo}': {e}")
    return dados
```

6. Escreva um script que importe o módulo criado acima e faça um histograma para cada coluna do arquivo, com seus respectivos valores. Use o dicionário ou a estrutura de dados criada em 1).

```python
import matplotlib.pyplot as plt
import numpy as np
# Importa a função de carregamento de dados do módulo

from data_loader_module import ler_arquivos_com_delimiter

def criar_histogramas_para_colunas(nome_arquivo, delimiter=','):
    """
    Carrega dados de um arquivo e gera um histograma para cada coluna.
    Assume que os dados são numéricos para plotagem.
    """
    dados = ler_arquivos_com_delimiter(nome_arquivo, delimiter=delimiter)

    if not dados:
        print("Não foi possível carregar os dados ou o arquivo está vazio. Abortando.")
        return

    print(f"Dados carregados com sucesso. Colunas encontradas: {list(dados.keys())}")

    # Itera sobre cada coluna no dicionário de dados
    for nome_coluna, valores in dados.items():
        try:
            # Tenta converter os valores da coluna para números (float)
            valores_numericos = [float(v) for v in valores]

            # Cria o histograma usando matplotlib
            plt.figure(figsize=(8, 5))
            plt.hist(valores_numericos, bins=10, color='skyblue', edgecolor='black')

            # Adiciona títulos e rótulos
            plt.title(f'Histograma da Coluna: {nome_coluna}')
            plt.xlabel('Valor')
            plt.ylabel('Frequência')
            plt.grid(axis='y', alpha=0.75)

            # Exibe o gráfico para a coluna atual
            plt.show()

        except ValueError:
            # Caso a coluna contenha dados não numéricos (como nomes)
            print(f"Pulando a coluna '{nome_coluna}': Contém dados não numéricos que não podem ser plotados em um histograma.")
        except Exception as e:
            print(f"Ocorreu um erro ao processar a coluna '{nome_coluna}': {e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Importa a função de carregamento de dados do módulo
from data_loader_module import ler_arquivos_com_delimiter

def analisar_estatisticas_coluna(valores_numericos, nome_coluna):
    """
    Calcula e imprime a média, desvio padrão amostral e desvio padrão populacional
    para uma lista de valores numéricos.
    """
    # Converte a lista Python em um array NumPy para cálculos otimizados
    dados_np = np.array(valores_numericos)

    # Calcula as estatísticas
    media = np.mean(dados_np)
    # ddof=1 especifica Graus de Liberdade de 1, para desvio padrão amostral
    desvio_padrao_amostral = np.std(dados_np, ddof=1)
    # ddof=0 (padrão) especifica Graus de Liberdade de 0, para desvio padrão populacional
    desvio_padrao_populacional = np.std(dados_np, ddof=0)

    print(f"\n--- Estatísticas para '{nome_coluna}' ---")
    print(f"Média: {media:.4f}")
    print(f"Desvio Padrão Amostral: {desvio_padrao_amostral:.4f}")
    print(f"Desvio Padrão Populacional: {desvio_padrao_populacional:.4f}")
    print("-" * (len(nome_coluna) + 25))


def criar_histogramas_para_colunas(nome_arquivo, delimiter=','):
    """
    Carrega dados de um arquivo, gera um histograma e imprime estatísticas para cada coluna.
    """
    dados = ler_arquivos_com_delimiter(nome_arquivo, delimiter=delimiter)

    if not dados:
        print("Não foi possível carregar os dados ou o arquivo está vazio. Abortando.")
        return

    print(f"Dados carregados com sucesso. Colunas encontradas: {list(dados.keys())}")

    # Itera sobre cada coluna no dicionário de dados
    for nome_coluna, valores in dados.items():
        try:
            # Tenta converter os valores da coluna para números (float)
            valores_numericos = [float(v) for v in valores]

            # CHAMA A NOVA FUNÇÃO DE ANÁLISE ESTATÍSTICA AQUI
            analisar_estatisticas_coluna(valores_numericos, nome_coluna)

            # Cria o histograma usando matplotlib
            plt.figure(figsize=(8, 5))
            plt.hist(valores_numericos, bins=10, color='skyblue', edgecolor='black')

            # Adiciona títulos e rótulos
            plt.title(f'Histograma da Coluna: {nome_coluna}')
            plt.xlabel('Valor')
            plt.ylabel('Frequência')
            plt.grid(axis='y', alpha=0.75)

            # Exibe o gráfico para a coluna atual
            plt.show()

        except ValueError:
            # Caso a coluna contenha dados não numéricos (como nomes)
            print(f"\nPulando a coluna '{nome_coluna}': Contém dados não numéricos que não podem ser plotados ou analisados estatisticamente.")
        except Exception as e:
            print(f"\nOcorreu um erro ao processar a coluna '{nome_coluna}': {e}")
```
