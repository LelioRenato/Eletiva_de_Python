2. Crie uma função que calcule e imprima velocidade media de um objeto a partir de uma posição inicial, a final e o tempo transcorrido para um objeto em MRU. Também crie uma funcão que calcule e imprima a velocidade de um objeto a partir da aceleração constante e o tempo (MRUA) (p.ex. queda libre).

```python
# Função para velocidade média em MRU
def v_mru(p_inicial, p_final, t):
    v = (p_final - p_inicial) / t
    print(f"Velocidade média (MRU): {v:.2f} m/s")

# Função para velocidade em MRUA (queda livre, etc.)
def v_mrua(aceleracao, t, v0=0):
    v = v0 + aceleracao * t
    print(f"Velocidade (MRUA): {v:.2f} m/s")
```

3. Crie uma funcão para calcular o ángulo zenital do sol (da semana passada) tomando como argumento as medidas da altura e o comprimento da sombra.

```python
import math

def calcular_angulo_zenital(altura_objeto, comprimento_sombra):
    """
    Calcula o ângulo zenital do sol em graus a partir da altura do objeto
    e do comprimento da sombra.

    Argumentos:
    altura_objeto: A altura do objeto (em qualquer unidade).
    comprimento_sombra: O comprimento da sombra (na mesma unidade da altura).

    Retorna:
    O ângulo zenital em graus.
    """
    if comprimento_sombra == 0:
        # Se a sombra for zero, o sol está diretamente acima (zênite), ângulo zenital é 0.
        return 0.0

    # Calculando o ângulo zenital em radianos usando a função atan (arco tangente)
    # Sendo a tangente do ângulo zenital: (comprimento_sombra / altura_objeto)
    angulo_zenital_radianos = math.atan(comprimento_sombra / altura_objeto)

    # Convertendo de radianos para graus
    angulo_zenital_graus = math.degrees(angulo_zenital_radianos)

    return angulo_zenital_graus
```

4. Crie uma função que faça a conversão de uma medida inicialmente em milhas para m, e outra para o inverso; uma de horas para segundos, e o inverso. Utilize estas funções para resolver novamente o primeiro exercício da semana passada (da corrida). Se uma pessoa demora 30 minutos em 4 milhas, qual velocidade media em km/h ? e o tempo medio por kilometro?

```python
# Funções de conversão
def milhas_para_m(milhas):
    """Converte milhas para metros."""
    return milhas * 1609.344

def m_para_milhas(metros):
    """Converte metros para milhas."""
    return metros / 1609.344

def horas_para_segundos(horas):
    """Converte horas para segundos."""
    return horas * 3600

def segundos_para_horas(segundos):
    """Converte segundos para horas."""
    return segundos / 3600

# Dados do problema
distancia_milhas = 4
tempo_minutos = 30

# Convertendo distância de milhas para km
distancia_m = milhas_para_m(distancia_milhas)
distancia_km = distancia_m / 1000

# Convertendo tempo de minutos para horas
tempo_horas = tempo_minutos / 60

# Calculando a velocidade média em km/h
velocidade_media_kmh = distancia_km / tempo_horas

# Calculando o tempo médio por quilômetro em minutos
tempo_medio_por_km =  tempo_minutos / distancia_km

# Resposta
print(f"velocidade_media_kmh: {velocidade_media_kmh:.2f}km/h")
print(f"tempo_medio_por_km: {tempo_medio_por_km:.2f} min")
```

5. Crie funções para calcular os outros exemplos das aulas anteriores: IMC, volume de uma esfera, distancia entre pontos de máximos de difração. Decida quais serão os argumentos e o valor retornado.

```python
import math

# Função para calcular o Índice de Massa Corporal (IMC)
def calcular_imc(peso_kg, altura_m):
    """
    Calcula o Índice de Massa Corporal (IMC).

    Argumentos:
    peso_kg: O peso da pessoa em quilogramas (kg).
    altura_m: A altura da pessoa em metros (m).

    Retorna:
    O valor do IMC calculado.
    """
    if altura_m <= 0:
        return None # Evita divisão por zero
    imc = peso_kg / (altura_m ** 2)
    return imc

# Função para calcular o volume de uma esfera
def calcular_volume_esfera(raio):
    """
    Calcula o volume de uma esfera.

    Argumento:
    raio: O raio da esfera.

    Retorna:
    O volume da esfera calculado.
    """
    if raio < 0:
        return None # O raio não pode ser negativo
    volume = (4/3) * math.pi * (raio ** 3)
    return volume

# Função para calcular a distância entre máximos de difração
def calcular_distancia_difracao(comprimento_onda, distancia_tela, fenda_separacao):
    """
    Calcula a distância entre os pontos de máximos de difração adjacentes.

    Argumentos:
    comprimento_onda: O comprimento de onda da luz (lambda).
    distancia_tela: A distância da fenda até a tela (L).
    fenda_separacao: A separação entre as fendas (d).

    Retorna:
    A distância entre máximos adjacentes (delta_y).
    """
    if fenda_separacao == 0:
        return None # Evita divisão por zero
    distancia_maximos = (comprimento_onda * distancia_tela) / fenda_separacao
    return distancia_maximos
```
