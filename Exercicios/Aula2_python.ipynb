Exercício 4:

Um laser vermelho (com comprimento de onda 632.8 nm) incide em uma fenda dupla produzindo um padrão de interferência com franjas claras e escuras, em um anteparo situado a uma distância D = 1.98 m da fenda. Calcule a distância
delta y entre dois máximos consecutivos de interferência. Considere o espaçamento entre as fendas, d , como sendo igual a 0.250 mm. Dica: a distância entre dois máximos de interferência consecutivos pode ser aproximada por delta y igual a lambida D/ d.

```python
#Dados

lam = 632.8e-9   # em m
D = 1.98         # em m
d = 0.250e-3     # em m

delta_y = lam * D / d
print(f"Δy = {delta_y:.6e} m = {delta_y*1e3:.3f} mm")
```
