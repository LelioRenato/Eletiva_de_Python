1. Utilizando dicionários e o método `get` dos dicionários, contrua a seguinte matriz em Python e acesse os valores de índices (0,3), (2,1), (4,3) e (2,2).

```python
# A matriz é representada como um dicionário onde cada chave (índice da linha) contém outro dicionário (índices das colunas e valores)

matriz = {
    0: {0: 1, 1: 2, 2: 3, 3: 4},
    1: {0: 5, 1: 6, 2: 7, 3: 8},
    2: {0: 9, 1: 0, 2: 1, 3: 2},
    3: {0: 3, 1: 4, 2: 5, 3: 6},
    4: {0: 7, 1: 8, 2: 9, 3: 0},
}

# Acessando os valores solicitados usando o método .get()

# Acessando o valor no índice (0, 3)
# Usando o .get(linha, {}) para garantir que a linha exista antes de tentar acessar a coluna.
valor_0_3 = matriz.get(0, {}).get(3, "Índice Inexistente")
print(f"Valor no índice (0, 3): {valor_0_3}")

# Acessando o valor no índice (2, 1)
valor_2_1 = matriz.get(2, {}).get(1, "Índice Inexistente")
print(f"Valor no índice (2, 1): {valor_2_1}")

# Acessando o valor no índice (4, 3)
valor_4_3 = matriz.get(4, {}).get(3, "Índice Inexistente")
print(f"Valor no índice (4, 3): {valor_4_3}")

# Acessando o valor no índice (2, 2)
valor_2_2 = matriz.get(2, {}).get(2, "Índice Inexistente")
print(f"Valor no índice (2, 2): {valor_2_2}")
```

2. Contagem de Letras.

  a.  Utilizando dicionários, escreva uma função que tome como argumento uma palavra (string) e  retorne um dicionário que tenha como ***key*** cada  letra da  palavra e como ***valor*** o número de vezes que ela aparece na palavra. Tal dicionário é chamado de **tabela de frequência**. ***Dica***: Utilize novamente o método `get` dos dicionários.

  b.  Utilizando os métodos `sort` (referente a listas) e `items` (referente a dicionários), imprima na tela uma lista ordenada alfabeticamente da tabela de frequência obtida no item a.

```python
# a

def criar_tabela_frequencia(palavra):
    """
    Cria uma tabela de frequência para uma palavra (string).

    Args:
        palavra (str): A palavra de entrada.

    Returns:
        dict: Um dicionário onde as chaves são as letras e os valores são
              as contagens de cada letra.
    """
    # Inicializa um dicionário vazio para armazenar as contagens
    tabela_frequencia = {}

    # Itera sobre cada letra na palavra de entrada
    for letra in palavra:
        # O método get() verifica se a letra já é uma chave no dicionário.
        # Se for, ele retorna o valor atual (a contagem), se não for, ele retorna o valor padrão 0.
        # Adicionando 1 a esse valor e atualizamos o dicionário.
        contagem_atual = tabela_frequencia.get(letra, 0)
        tabela_frequencia[letra] = contagem_atual + 1

    return tabela_frequencia

#Exemplo
palavra_exemplo = "programacao"
frequencia = criar_tabela_frequencia(palavra_exemplo)

print(f"Palavra original: '{palavra_exemplo}'")
print(f"Tabela de frequência: {frequencia}")

palavra_banana = "banana"
frequencia_banana = criar_tabela_frequencia(palavra_banana)
print(f"\nPalavra original: '{palavra_banana}'")
print(f"Tabela de frequência: {frequencia_banana}")
```

```python
# b

def criar_tabela_frequencia(palavra):
    """
    Cria uma tabela de frequência para uma palavra (string).
    """
    tabela_frequencia = {}
    for letra in palavra:
        contagem_atual = tabela_frequencia.get(letra, 0)
        tabela_frequencia[letra] = contagem_atual + 1
    return tabela_frequencia

def imprimir_frequencia_ordenada(frequencia_dict):
    """
    Imprime a tabela de frequência ordenada alfabeticamente pelas letras (chaves).
    """
    # Obter uma lista de pares (letra, contagem) usando .items()
    # Exemplo: de {'b': 3, 'a': 3, 'n': 2} para [('b', 3), ('a', 3), ('n', 2)]
    lista_itens = list(frequencia_dict.items())

    # Ordenar a lista usando .sort()
    # O método sort() ordena a lista in-place (diretamente), por padrão, ele ordena pelo primeiro elemento do par (a letra).
    lista_itens.sort()

    print("\nTabela de frequência ordenada alfabeticamente:")
    for letra, contagem in lista_itens:
        print(f"'{letra}': {contagem}")

# Exemplo de Uso Integrado

palavra_exemplo = "ornitorrinco"

# Item a: Criação da tabela de frequência
frequencia = criar_tabela_frequencia(palavra_exemplo)
print(f"Tabela original (não ordenada): {frequencia}")

# Item b: Impressão da lista ordenada
imprimir_frequencia_ordenada(frequencia)
```

3. Escreva um programa que leia uma string e retorne uma tabela das letras do alfabeto em ordem alfabética que ocorrem na string junto com o número de vezes que cada letra ocorre. Deve ser ignorado se a letra ocorre em maiúscula ou minúscula (contar da mesma forma "A" e "a"). Um exemplo de saída do programa quando o usuário digita os dados “ThiS is String with Upper and lower case Letters”, seria o seguinte:

a  2

c  1


d  1


e  5


g  1


h  2


i  4


l  2


n  2


o  1


p  2


r  4


s  5


t  5


u  1


w  2

```python
import string

def contar_letras_ordenado(texto):
    """
    Lê uma string e retorna uma tabela de frequência ordenada alfabeticamente
    para as letras do alfabeto que ocorrem na string.
    """
    # Preparação dos dados
    # Converte o texto inteiro para minúsculas para contar 'A' e 'a' juntos
    texto = texto.lower()

    # Obtém todas as letras minúsculas do alfabeto
    alfabeto = string.ascii_lowercase

    # Dicionário para armazenar a frequência
    frequencia = {}

    # Contagem das letras
    for letra in texto:
        # Verifica se o caractere é realmente uma letra do alfabeto antes de contar
        if letra in alfabeto:
            # Usando o método .get() para contar
            frequencia[letra] = frequencia.get(letra, 0) + 1

    # Ordenação e Impressão (usando os métodos items() e sort() implicitamente)
    # Como iteramos sobre o texto e as letras do alfabeto são inseridas na ordem que aparecem pela primeira vez.

    print("\n--- Tabela de Frequência Alfabética ---")

    for letra_alfabeto in alfabeto:
        # Verifica se a letra do alfabeto existe nos dados que contamos
        if letra_alfabeto in frequencia:
            contagem = frequencia[letra_alfabeto]
            print(f"{letra_alfabeto}  {contagem}")

# Teste

# Solicita a entrada do usuário
entrada_usuario = input("Digite uma frase ou palavra: ")

# Chama a função para processar e exibir o resultado
contar_letras_ordenado(entrada_usuario)
```
