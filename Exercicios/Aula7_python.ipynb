#O loop for revisitado

Podemos atualizar uma variável dentro de um loop for.

Para tal, vamos escrever uma função que calcula a soma de todos os ítens de uma lista.

```python
def mysum(xs):
    """ Soma todos os números em uma dada lista, e retorna o total. """
    total = 0                # inicialização da variável total
    for x in xs:             # para cada item da lista xs \ começo do loop
        total += x           # somar o valor anterior ao item atual da lista: total = total + x
#        print(total)         # serve para ver os valores da lista se estão somando
    return total

#print(mysum([2,4,6,8]))    # comentado, ainda mesmo que não apareça nada, significa que ele aceitou os valores da lista

assert mysum([2, 4, 6, 8])  # declaração para testar a validade das condições da função mysum. Veremos mais a seguir.
```

#A declaração while

```python
def soma_ate(n):
    """ Retorna a soma de 1+2+3 ... n """  # n é um numero inteiro
    ss  = 0                 # inicializo para zero
    v = 1                   # inicializo para um
    while v <= n:           # essa é a condição, enquanto for VERDADEIRA, vai continuando executando!
        ss = ss + v     # ss += v
        print('ss = ', ss)
        v = v + 1       # v +=1 (forma mais resumida, sempre quando ver algo assim), precisa para atualizar v, pois, é necessário para continuar verdadeiro.
        print('v = ', v)
    return ss

print(soma_ate(8))
```

#Contadores

Para contar os digitos de um dado número n, podemos escrever uma função:

```python
def counter(n):             # parametro n
    count = 0               # inicialização do contador, alem disso, count é uma variavel local (definido localmente em uma função)
    while n!=0:             # condição para o loop ser satisfeita!
        count = count + 1
        n = n // 10         # divisão de inteiros, ficando só a parte inteira
    return count

counter(7100)
```

#A declaração break

A instrução break é usada para deixar imediatamente o corpo do loop. A próxima instrução a ser executada é a primeira após o corpo:

```python
for i in [12, 16, 17, 24, 29]:
    if i % 2 == 1:  # If the number is odd, ou seja, se for igual a 1 ele sai do loop (numero impar)
       break        #  ... immediately exit the loop, ou seja, vai imprimir os itens até o primeiro sair (17)
    print(i)
print("done")
```

#A declaração continue

Esta é uma instrução de fluxo de controle que faz com que o programa pule imediatamente o processamento do resto do corpo do loop, para a iteração atual. Mas o loop ainda continua correndo pelas iterações restantes:

```python
# Mesma coisa do break, porem, aqui ele continua em vez de parar

for i in [12, 16, 17, 24, 29, 30]:
    if i % 2 == 1:      # If the number is odd, ou seja, o que for impar ele vai pular em vez de imprimir
       continue         # Don't process it
    print(i)
print("done")
```

2. Modifique a função counter para contar o número de vezes que os dígitos 0 e 5 aparecem em um dado número n. Por exemplo, n = 10568, tem 2 dígitos 0 ou 5.

```python
def counter(n):
    """
    Conta o número de vezes que os dígitos 0 e 5 aparecem em um número.

    Argumento:
    n: O número a ser verificado.

    Retorna:
    O número de ocorrências dos dígitos 0 e 5.
    """
    contador = 0
    # Converte o número para uma string para que possa ser iterado dígito por dígito
    s = str(n)

    # Itera sobre cada caractere da string
    for digito in s:
        # Verifica se o dígito atual é '0' ou '5'
        if digito == '0' or digito == '5':
            contador += 1

    return contador

#Exemplo

n1 = 10568
print(f"O número {n1} tem {counter(n1)} dígitos 0 ou 5.")
```

4. O método de Newton para raiz quadrada. Suponha que você queira calcular a raiz quadrada de n. Se você começar com quase qualquer aproximação, poderá calcular uma aproximação melhor (mais próxima da resposta real) com a seguinte fórmula: melhor = (approx + n/approx)/2
Usando um loop e repetindo esta fórmula até que a melhor aproximação se aproxime o suficiente da anterior (este "suficiente" é arbitrario: podemos considerar uma diferença de 0,001), escreva uma função para calcular a raiz quadrada utilizando o método de Newton.

Esse é um exemplo de repetição indefinida. Com qual tipo de loop você deve escrever o seu programa?

```python
def newton_raiz_quadrada(n):
    """
    Calcula a raiz quadrada de n usando o método de Newton.
    A iteração continua até que a diferença entre as aproximações seja menor que 0.001.
    """
    # Condição inicial: "ponto de partida".
    # Usando n/2 ou 1.0 (para começar).
    aproximacao_anterior = n / 2.0
    diferenca = 1000.0 # Inicializando com um valor grande para garantir que o loop while inicie

    # Definindo a tolerância especificada
    tolerancia = 0.001

    # O loop continua ENQUANTO a diferença entre as aproximações for maior que a tolerância
    while diferenca >= tolerancia:
        # Calcula a 'melhor' nova aproximação usando a fórmula de Newton
        melhor_aproximacao = (aproximacao_anterior + n / aproximacao_anterior) / 2.0

        # Calcula a diferença absoluta entre a nova e a velha aproximação
        diferenca = abs(melhor_aproximacao - aproximacao_anterior)

        # Atualiza a aproximação anterior para a próxima iteração
        aproximacao_anterior = melhor_aproximacao

    # Quando o loop termina, a última 'melhor_aproximacao' será a resposta final
    return melhor_aproximacao

#Teste

numero_para_calcular = 25
raiz = newton_raiz_quadrada(numero_para_calcular)
print(f"A raiz quadrada de {numero_para_calcular} é aproximadamente: {raiz:.4f}")

numero_para_calcular_2 = 12345
raiz_2 = newton_raiz_quadrada(numero_para_calcular_2)
print(f"A raiz quadrada de {numero_para_calcular_2} é aproximadamente: {raiz_2:.4f}")
```

```python
import time
time.time()
```

Escreva um script que leia a hora atual e a converta em um tempo em horas, minutos e segundos, mais o número de dias desde a época.

```python
import time

def converter_tempo_epoca():
    """
    Lê a hora atual do sistema e a converte em dias, horas, minutos e segundos
    desde a época (1 de janeiro de 1970).
    """
    #Obter o número total de segundos desde a época
    segundos_totais = time.time()
    print(f"Segundos totais desde a época: {segundos_totais:.2f}")

    # Calcular o número de dias
    # Há 86400 segundos em um dia (60 seg * 60 min * 24 horas), poderia ter feito uma argumento para isso mas fiquei com preguiça
    segundos_por_dia = 86400
    dias_desde_epoca = segundos_totais // segundos_por_dia

    # Calcular as horas, minutos e segundos restantes
    # O operador '%' (módulo) nos dá o restante da divisão por dias
    segundos_restantes = segundos_totais % segundos_por_dia

    horas = segundos_restantes // 3600
    segundos_restantes %= 3600

    minutos = segundos_restantes // 60
    segundos_finais = segundos_restantes % 60

    #Imprimindo o resultado
    print("-" * 30)
    print(f"Tempo detalhado desde a época:")
    print(f"Dias: {int(dias_desde_epoca)}")
    print(f"Horas: {int(horas)}")
    print(f"Minutos: {int(minutos)}")
    print(f"Segundos: {segundos_finais:.2f}")
    print("-" * 30)

# Executa a função principal
converter_tempo_epoca()
```

```python
def draw(t, length, n):
    '''
    t: turtle ou ColabTurtle
    length: comprimento da linha
    n: número de iterações. deve ser maior que zero.
    '''
    if n == 0:
        return
    angle = 50
    t.forward(length * n)
    t.left(angle)
    draw(t, length, n-1)
    t.right(2 * angle)
    draw(t, length, n-1)
    t.left(angle)
    t.backward(length * n)
```

```python
import turtle
import time

def draw(t, length, n):
    '''
    t: turtle object
    length: comprimento da linha base
    n: número de iterações/níveis de recursão.
    '''
    if n == 0:
        return
    angle = 50
    t.forward(length * n)
    t.left(angle)
    draw(t, length, n-1)  # Chama recursivamente o ramo esquerdo
    t.right(2 * angle)
    draw(t, length, n-1)  # Chama recursivamente o ramo direito
    t.left(angle)
    t.backward(length * n)

# --- Configuração para executar o desenho ---
jn = turtle.Screen()
jn.bgcolor("white")
madara = turtle.Turtle()
madara.color("green")
madara.pensize(2)
madara.speed(0)    # Coloca velocidade máxima

# Posiciona a tartaruga na parte inferior da tela, apontando para cima
madara.penup()
madara.goto(0, -200)
madara.left(90)
madara.pendown()

# Executando a função
# Desenhando uma "árvore" com 4 níveis de recursão (n=4) e comprimento base 5
draw(madara, 5, 4)
```

5. A curva de Koch é um fractal que parece com o da Figura abaixo.

Para desenhar uma curva de Koch com o comprimento x, tudo o que você tem que fazer é:

1. Desenhe uma curva de Koch com o comprimento x/3.

1. Vire 60 graus à esquerda.

1. Desenhe uma curva de Koch com o comprimento x/3.

1. Desenhe uma curva de Koch com o comprimento x/3.

A exceção é se x for menor que 3: neste caso, você pode desenhar apenas uma linha reta com o comprimento x.

1. Escreva uma função chamada `koch` que receba um `turtle` (usar [trinket](https://trinket.io/turtle)) e um comprimento como parâmetros, e use o `turtle` para desenhar uma curva de Koch com o comprimento dado.

1. Escreva uma função chamada `snowflake` que desenhe três curvas de Koch para fazer o traçado de um floco de neve.

1. A curva de Koch pode ser generalizada de vários modos. Veja exemplos em http://en.wikipedia.org/wiki/Koch_snowflake e implemente o seu favorito.

```python
import turtle

def koch(t, x):
    """
    Desenha uma curva de Koch de comprimento x na Turtle t.
    """
    if x < 3:
        # Condição de parada (caso base): desenha uma linha reta curta
        t.forward(x)
    else:
        # Caso recursivo: aplica as 4 etapas fractais

        # 1. Desenhe uma curva de Koch com o comprimento x/3
        koch(t, x/3)

        # 2. Vire 60 graus à esquerda
        t.left(60)

        # 3. Desenhe uma curva de Koch com o comprimento x/3
        koch(t, x/3)

        # 4. Vire 120 graus à direita (ângulo externo da ponta)
        t.right(120)

        # 5. Desenhe uma curva de Koch com o comprimento x/3
        koch(t, x/3)

        # 6. Vire 60 graus à esquerda
        t.left(60)

        # 7. Desenhe uma curva de Koch com o comprimento x/3
        koch(t, x/3)

def snowflake(t, x):
    """
    Desenha um floco de neve de Koch (triângulo de Koch) desenhando
    três curvas de Koch, girando 120 graus à direita entre cada uma.
    """
    # Para desenhar a primeira curva, ajustamos a posição inicial
    t.penup()
    t.goto(-x/2, x/3) # Posiciona a tartaruga para que o floco fique centralizado
    t.pendown()

    # Desenha os três lados do floco de neve
    for _ in range(3):
        koch(t, x)
        t.right(120) # Gira 120 graus para o próximo lado do triângulo

#Configuração e execução do programa

if __name__ == '__main__':
    jn = turtle.Screen()
    jn.bgcolor("white")
    naruto = turtle.Turtle()
    naruto.color("blue")
    naruto.pensize(2)
    naruto.speed(0)

    # Desenha o floco de neve com comprimento total de lado 300
    snowflake(naruto, 300)

    naruto.hideturtle() # Esconde a tartaruga no final
    jn.mainloop()
```
