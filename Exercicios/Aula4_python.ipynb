1. Escreva uma função com um valor de retorno que:

tome como input dois argumentos;

execute a soma dos dois parâmetros que foram passados como argumento e atribua o resultado a uma variável;

tenha como valor de retorno o dobro da soma dos dois parâmetros.

```python
def calcular_dobro_da_soma(numero1, numero2):
    """
    Calcula a soma de dois números e retorna o dobro dessa soma.

    Argumentos:
    numero1: O primeiro número.
    numero2: O segundo número.

    Retorna:
    O dobro da soma dos dois números.
    """
    # Executa a soma dos dois parâmetros e atribui o resultado a uma variável
    soma = numero1 + numero2

    # Retorna o dobro da soma dos parâmetros
    resultado_final = soma * 2
    return resultado_final
```
