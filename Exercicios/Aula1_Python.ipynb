Exercício 3:

Inicie o interpretador Python e use-o como uma calculadora. A sintaxe do Python para operações matemáticas é quase igual à notação matemática padrão:

adição: +
subtração: -
divisão: /
multiplicação: *

Se você fizer uma corrida de 10 quilômetros em 43 minutos e 30 segundos, qual será seu tempo médio por milha? Qual é a sua velocidade média em milhas por hora? (Dica: há 1,61 quilômetros em uma milha).

```python
# Dados
d_km = 10  #distancia
t_min = 43  #tempo em minuto
t_seg = 30  # tempo em segundo

# Conversões
km_por_milha = 1.61
d_milhas = d_km / km_por_milha

t_total_horas = (t_min * 60 + t_seg) / 3600

# Cálculos
t_por_milha_horas = t_total_horas / d_milhas
v_medio = d_milhas / t_total_horas  #velocidade média

# Converter tempo médio por milha para min e seg
t_por_milha_min = int(t_por_milha_horas * 60)
t_por_milha_seg = int((t_por_milha_horas * 60 - t_por_milha_min) * 60)

# Resultados
print(f"Distância em milhas: {d_milhas:.2f}")
print(f"Tempo médio por milha: {t_por_milha_min} min {t_por_milha_seg} s")
print(f"Velocidade média: {v_medio:.2f} mph")
```
