1. Percorra toda a string "banana", utilizando as iterações apresentadas nas aulas anteriores, e imprima na tela uma letra por linha como se fosse em uma escada.

```python
# For

palavra = "banana"

print("--- Iteração direta na string ---")
for letra in palavra:
    print(letra)
```

```python
# For com range

palavra = "banana"
tamanho_da_palavra = len(palavra)

print("--- Iteração usando range() e índices ---")
for i in range(tamanho_da_palavra):
    # Acessa a letra na posição i
    print(palavra[i])
```

```python
# While

palavra = "banana"
indice = 0
tamanho_da_palavra = len(palavra)

print("--- Iteração usando loop while ---")
while indice < tamanho_da_palavra:
    print(palavra[indice])
    indice += 1
```

2. Escreva uma função que receba uma string como argumento e exiba as letras de trás para a frente, uma por linha.

```python
def exibir_reverso_slice(texto):
    """
    Recebe uma string e exibe suas letras de trás para frente, uma por linha,
    usando a notação de slice reverso.
    """
    # Reverte a string: [início:fim:passo], sendo que o passo -1 inverte a string.
    texto_revertido = texto[::-1]

    # Itera sobre a string revertida e imprime cada caractere
    for letra in texto_revertido:
        print(letra)

#Teste
palavra_teste = "Pegasus Seiya"
print(f"Exibindo '{palavra_teste}' de trás para frente:")
exibir_reverso_slice(palavra_teste)
```

1. Resolução numérica de equações diferenciais. Imagine que tem um/a maratonista treinando, que num trecho de  5km  tem velocidade quase constante de  12km/h . Produza uma lista com a distância percorrida em intervalos de 1 minuto.

```python
# Dados do problema
velocidade_km_h = 12  # Velocidade em km/h
distancia_total_km = 5  # Distância total a ser percorrida
intervalo_minutos = 1  # Intervalo de tempo para cada medição

# Convertendo a velocidade de km/h para km/min
velocidade_km_min = velocidade_km_h / 60

#Simulando o evento
distancias_percorridas = []
distancia_acumulada = 0
tempo_decorrido_minutos = 0

# O loop continua enquanto a distância acumulada for menor que a distância total
while distancia_acumulada < distancia_total_km:
    # Adiciona o intervalo de tempo
    tempo_decorrido_minutos += intervalo_minutos

    # Adiciona a distância percorrida no último minuto
    distancia_acumulada += velocidade_km_min

    # Adiciona a nova distância acumulada à lista
    # Usando o round() para formatar o número e evitar imprecisões de ponto flutuante
    distancias_percorridas.append(round(distancia_acumulada, 2))

#Exibindo a lista
print("Distância percorrida em intervalos de 1 minuto:")
for minuto, distancia in enumerate(distancias_percorridas, 1):
    print(f"Minuto {minuto}: {distancia} km")

print("\n--- Resumo ---")
print(f"O maratonista completou {distancia_total_km} km em {tempo_decorrido_minutos} minutos.")
```

2. O/A mesmo/a atleta no seguinte trecho acelera nos seguintes  200  metros até chegar numa velocidade de  15km/h , e mantém esta velocidade por  1,8km . Queremos obter uma tabela ou gráfico da distância percorrida en função do tempo, e calcular e o tempo total para chegar em  7km , faça isto utilizando listas.

```python
import matplotlib.pyplot as plt

def simular_trajeto():
    """
    Simula o trajeto do atleta e retorna listas de tempo e distância.
    """
    # Constantes
    distancia_total_objetivo_km = 7.0
    intervalo_tempo_segundos = 1 # Para um melhor resultado

    # Listas para armazenar os resultados (dados fornecidos)
    tempos = []
    distancias = []

    # Variáveis de estado iniciais
    tempo_acumulado_h = 0.0
    distancia_acumulada_km = 0.0

    # Converter unidades base para km/h e horas, usando o passo de 1 segundo (1/3600 horas)
    dt_h = intervalo_tempo_segundos / 3600.0

    # Parte 1: Aceleração (200m = 0.2km), de 12 km/h para 15 km/h ---
    v_inicial_kmh = 12.0
    v_final_kmh = 15.0
    distancia_seg1_km = 0.2

    #Calculando o tempo total do segmento e dividindo os dados
    tempo_total_seg1_h = distancia_seg1_km / ((v_inicial_kmh + v_final_kmh) / 2)
    passos_seg1 = int(tempo_total_seg1_h / dt_h)

    print(f"Parte 1: Acelerando por {distancia_seg1_km*1000:.0f}m em {tempo_total_seg1_h*3600:.1f} segundos.")

    for i in range(passos_seg1):
        # Velocidade instantânea aumenta linearmente
        velocidade_atual = v_inicial_kmh + (v_final_kmh - v_inicial_kmh) * (i / passos_seg1)
        distancia_acumulada_km += velocidade_atual * dt_h
        tempo_acumulado_h += dt_h
        tempos.append(tempo_acumulado_h * 60) # Armazena em minutos para o gráfico
        distancias.append(distancia_acumulada_km)

    # Parte 2: Velocidade Constante (1.8km a 15 km/h)
    v_constante_kmh = 15.0
    distancia_seg2_km = 1.8
    distancia_alvo_seg2 = distancia_acumulada_km + distancia_seg2_km

    print(f"Parte 2: Mantendo {v_constante_kmh} km/h por {distancia_seg2_km} km.")

    while distancia_acumulada_km < distancia_alvo_seg2 - (v_constante_kmh * dt_h):
        distancia_acumulada_km += v_constante_kmh * dt_h
        tempo_acumulado_h += dt_h
        tempos.append(tempo_acumulado_h * 60)
        distancias.append(distancia_acumulada_km)

    # Ajuste final para garantir o ponto exato de 7km se necessário
    if distancias[-1] < distancia_total_objetivo_km:
        distancia_restante = distancia_total_objetivo_km - distancias[-1]
        tempo_extra_h = distancia_restante / v_constante_kmh
        tempo_acumulado_h += tempo_extra_h
        distancia_acumulada_km = distancia_total_objetivo_km
        tempos.append(tempo_acumulado_h * 60)
        distancias.append(distancia_acumulada_km)

    print(f"Parte 3: Tempo total para 7km calculado a partir do fim da Parte 2.")

    return tempos, distancias, tempo_acumulado_h

#Executando o código

tempos_min, distancias_km, tempo_total_horas = simular_trajeto()

# 1. Calculando o tempo total para chegar em 7km
tempo_total_minutos = tempo_total_horas * 60
print(f"\nTempo total para percorrer 7km: {tempo_total_minutos:.2f} minutos.")
print(f"Tempo total para percorrer 7km: {tempo_total_horas*3600:.2f} segundos.")

# 2. Gerando a Tabela para os primeiros e últimos 10 pontos
print("\nTabela de Distância vs Tempo (primeiros e últimos 10 pontos):")
print("Tempo (min) | Distância (km)")
print("-" * 30)
for i in range(10):
    print(f"{tempos_min[i]:>11.4f} | {distancias_km[i]:>14.4f}")
print("...")
for i in range(-10, 0):
    print(f"{tempos_min[i]:>11.4f} | {distancias_km[i]:>14.4f}")

# 3. Gerando o Gráfico (usando matplotlib)
plt.figure(figsize=(10, 6))
plt.plot(tempos_min, distancias_km, linestyle='-', color='b')
plt.title('Distância Percorrida em Função do Tempo')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Distância (km)')
plt.grid(True)
plt.axhline(y=7.0, color='r', linestyle='--', label='Objetivo 7km')
plt.axvline(x=tempo_total_minutos, color='g', linestyle='--', label=f'Tempo Total {tempo_total_minutos:.2f} min')
plt.legend()
plt.show()
```

3. Considere um péndulo simples num plano. O comprimento da linha é  1  metro, a massa é  0,2kg  e a aceleração de gravidade  9,8m/s2 . Faça um gráfico ou crie uma tabela para a posição da massa em coordenadas cartesianas em função do tempo, utilizando listas.

```python
import numpy as np
import matplotlib.pyplot as plt

#Parâmetros do Pêndulo
L = 1.0       # Comprimento da linha (metros)
g = 9.8       # Aceleração da gravidade (m/s^2)
m = 0.2       # Massa (kg)
theta_0 = np.radians(15) # Ângulo inicial (15 graus convertido para radianos)
omega_0 = 0.0 # Velocidade angular inicial (partindo do repouso)

#Parâmetros da Simulação
dt = 0.001    # Passo de tempo (segundos) - quanto menor, mais preciso será
tempo_total = 2.0 # Simulando por 2 segundos
num_passos = int(tempo_total / dt)

#Listas para armazenar os resultados
tempos = []
posicoes_x = []
posicoes_y = []

#Simulação Numérica (Método de Euler)
theta = theta_0
omega = omega_0
tempo = 0.0

for i in range(num_passos):
    # Calculando a aceleração angular (a = d^2(theta)/dt^2)
    alpha = -(g / L) * np.sin(theta)   # utilizando sin, para maior precisão

    # Atualizando a velocidade angular (omega) e posição angular (theta)
    omega += alpha * dt
    theta += omega * dt
    tempo += dt

    # Convertendo coordenadas polares (theta, L) para cartesianas (x, y)
    # Definido o ponto de pivô em (0, 0)
    x = L * np.sin(theta)
    y = -L * np.cos(theta)   # Negativo para que 'para baixo' seja positivo visualmente

    # Armazenar os dados nas listas
    tempos.append(tempo)
    posicoes_x.append(x)
    posicoes_y.append(y)

#Resultados

# 1. Tabela (mostrando os primeiros e últimos 10 pontos)
print("Tabela de Posição (X, Y) vs Tempo (primeiros e últimos 10 pontos):")
print("Tempo (s) | Posição X (m) | Posição Y (m)")
print("-" * 45)
for i in range(10):
    print(f"{tempos[i]:>9.4f} | {posicoes_x[i]:>13.4f} | {posicoes_y[i]:>13.4f}")
print("...")
for i in range(-10, 0):
    print(f"{tempos[i]:>9.4f} | {posicoes_x[i]:>13.4f} | {posicoes_y[i]:>13.4f}")

# 2. Gráfico da trajetória (usando matplotlib)
plt.figure(figsize=(6, 6))
plt.plot(posicoes_x, posicoes_y, label='Trajetória do Pêndulo')
plt.plot([0], [0], 'ko') # Ponto de pivô
plt.plot([posicoes_x[0]], [posicoes_y[0]], 'ro') # Posição inicial
plt.title('Trajetória Cartesiana do Pêndulo Simples')
plt.xlabel('Posição X (metros)')
plt.ylabel('Posição Y (metros)')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axis('equal') # Mantém a escala igual nos eixos para visualização correta do arco
plt.legend()
plt.grid(True)
plt.show()
```
